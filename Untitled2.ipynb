```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,color='blue',linestyle='--',marker='o',label='y=2x')
plt.title('Line Chart Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories=['A','B','C','D',]
values=[10,20,15,25]
plt.bar(categories,values,color='r',width=0.5)
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
x=[5,7,8,7,6,9,5,6]
y=[99,86,87,100,86,103,87, 90] # Added an extra value to match the length of x
plt.scatter(x,y,color='g',s=100,alpha=0.7)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
sizes = [20, 30, 25, 25]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
colors = ['gold','lightblue','lightgreen','pink']
explode = [0,0.1,0,0]  # Highlight Bananas

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title("Pie Chart Example")
plt.show()
```
