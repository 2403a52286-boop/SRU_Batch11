```python
import pandas as pd
import numpy as np
data={"name":['ricky','vicky','ricky','rocky'],
        "Age":[12,23,12,None],
       "place":['wgl','hnk','wgl',None]}
df=pd.DataFrame(data)
print("\n orignal data\n",df)
#handling missing values
df['Age']=df['Age'].fillna(df['Age'].mean())
df['place']=df['place'].fillna('usa')
print("\n data after handling missing values\n",df)
#removing duplicate values
df=df.drop_duplicates()
print("\n data after removing duplicate values\n",df)
#change data types
df['Age']=df['Age'].astype(int)
print("\n data after changing data types\n",df)
#remane the coloums
df=df.rename(columns={'name':'Name','place':'Place'})
print("\n data after renaming the coloums\n",df)
#filter the data
df=df[df['Age']>10]
print("\n data after filtering the data\n",df)
#sort the data
df=df.sort_values('Age',ascending=False)
print("\n data after sorting the data\n",df)
#group
group=df.groupby("Name").agg(AVG_Age=("Age","mean"),Total_Age=("Age","sum"))
print("\n data after grouping the data\n",group)
```
