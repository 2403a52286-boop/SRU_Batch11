```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'income_category': ['low', 'low', 'medium', 'high', 'medium', 'high', 'low', 'high', 'medium', 'low'],
    'loan_status': ['approved', 'rejected', 'approved', 'approved', 'rejected', 'rejected', 'approved', 'approved', 'approved', 'rejected'],
    'loan_amount': [5000, 3000, 20000, 35000, 25000, 45000, 7000, 40000, 22000, 6000]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
df.groupby(['income_category', 'loan_status']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Loan Status Distribution by Income Category')
plt.ylabel('Loan Count')
plt.xlabel('Income Category')
plt.xticks(rotation=0)
plt.show()

approval_rate = df.groupby('income_category')['loan_status'].value_counts(normalize=True).unstack().fillna(0)['approved']
plt.figure(figsize=(8, 6))
plt.plot(approval_rate.index, approval_rate.values, marker='o', color='b')
plt.title('Loan Approval Rate by Income Category')
plt.xlabel('Income Category')
plt.ylabel('Approval Rate')
plt.show()
```
